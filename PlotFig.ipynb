```python
import Jetson.GPIO as GPIO
import time

GPIO.cleanup()
GPIO.setmode(GPIO.BOARD)
GPIO.setup(40, GPIO.OUT)
```

```python
flag = False;
for i in range(15000):
    flag = not flag
    if flag == False:
        GPIO.output(40, GPIO.LOW)
        time.sleep(0.1)
    elif flag == True:
        GPIO.output(40, GPIO.HIGH)
        time.sleep(0.1)
```
